# 파일 쓰기(write)

* 1. 파일 열기(open)
    * `open(file, mode, ...)`
        * file: 폴더(디렉토리를 포함한 파일 이름
        * mode: 파일을 여는 방법(모드)
            * r(read) : 읽기. (기본값)
            * w(write) : 쓰기. 파일이 없으면 새로운 파일을 생성. 파일이 있으면 기존 내용이 모두 삭제되고 파일의 처음부터 다시 쓰기 시작.
            * a(append) : 이어쓰기. 기존 내용부터 이어서 쓰기.
            * t(text) : 텍스트 (읽기/ 쓰기). (기본값)
            * b(binary) : 이진 데이터 (읽기/쓰기). (예) 이미지, 동영상 등
* 2. 파일 내용 쓰기(write)
* 3. 파일 닫기(close)

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 1. 파일을 텍스트 쓰기 모드로 열기.
f = open('test1.txt', mode='wt')  # mode='w'와 동일

# 2. 열린 파일에 내용을 작성.
f.write('Hello, Python!\n')
f.write('안녕하세요...\n')
f.write('1234567890\n')

# 3. 파일 닫기.
f.close()
```

```python

# with ~ as ... 구문: 오픈된 리소스(파일)에서 close() 메서드를 자동으로 호출.
with open('test2.txt', mode='w') as f:
    f.write('안녕하세요...\n')
    f.write('Hello...\n')
```

#파일 읽기

```python

with open('test1.txt', mode='rt') as f:
    # mode='rt'가 기본값이기 때문에 생략 가능.
    line = f.readline()
    print(line.strip())
    line = f.readline()
    print(line.strip())
    line = f.readline()
    print(line.strip())
    
```

```python

with open('test1.txt') as f:
    for line in f:  # 파일의 처음부터 끝까지 readline() 자동으로 호출
        print(line.strip())
```

```python
dir_path = '/content/drive/MyDrive/Colab Notebooks/lab-python'
file_name = 'emails.txt'
```

```python
import os  # os 모듈을 임포트 - os.path.join() 함수를 사용하기 위해서
```

```python
# os.path.join(): 폴더 경로 + 파일 구분자 + 파일 이름
file_path = os.path.join(dir_path, file_name)
file_path
```
