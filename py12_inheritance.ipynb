# 상속(Inheritance)

* super class(상위 클래스), parent class(부모 클래스), base class(기본 클래스)
* sub class(하위 클래스), child class(자식 클래스), dericed class(유도 클래스)
* 상속 : 상위 클래스의 속성(데이터)들과 기능(메서드)들을 하위 클래스에서 재사용하는 것.
* 일반적으로 **IS-A** 관계가 성립하는 객체들을 상속을 사용해서 구현함.
    * 학생은 사람이다. Student IS A person
    * Percon - super class
    * Srudent - sub class
* **Has-A** 관계가 성립하는 객체들은 일반적으로 상속이 아니라 클래스의 **속성**(멤버 변수)로 구현
    * 학생은 성적을 갖고 있다.

## 파이썬에서 상속을 구현

```
Class SubClass(SuperClass):
    구현 코드
```

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
        # 변수 이름 앞에'__' 를 붙이면 private 변수가 됨

    def get_age(self):
        return self.__age

    def hello(self):
        print(f'안녕하세요. 저는 {self.get_age()}살 {self.name}입니다.')
````

````python
mrLee = Person('이존귤', 22)

# print(mrLee.__age) < 오류발생 : __age는 private 변수이기 때문에.

mrLee.hello()
````

````python
class Student(Person):
    pass
````

````python
stu1 = Student('이존귯', 20)

stu1.hello()
````

#Method Overeide

* 상위 클래스의 메서드를 하위 클래스에서 재정의 하는 것.
    * 파이썬에서는 메서드의 이름만 같으면 (파라미터 타입/ 개수와 상관 없이) 메서드가 재정의 됨.
        * 파이썬은 같은 이름으로 2개 이상의 메서드(함수)를 가질 수 없음 - Overloading은 제공하지 않음.
    * 자바에서는 메서드의 이름, 파라미터 타입/개수/순서가 모두 동일한 경우에 메서드가 재정의 됨.
        * 파라미터가 다르면 같은 이름의 메서드가 2개 이상 있을 수 있음 - overloading
* 하위 클래스에서 재정의(override)된 상위 클래스의 메서드를 명시적으로 호출하기 위해서는 `super(). method_name(...)` 의 형식으로 호출.


````python
class BuisnessPerson(Person):
    def __init__(self, name, age, company):
        #Super class의 __init__ 메서드를 호출
        super().__init__(name,age)
        self.company = company # 하위 클래스에서 상위 클래스를 확장한 속성.

    # 상위 클래스의 hello() 메서드를 override
    def hello(self):
        super().hello()
        print(f'저는 {self.company}에 근무합니다.')
````

````python
bui1 = BuisnessPerson('이존굿', 26, '샘플')

bui1.hello()

bui1.get_age()
````

````python
class Animal:
    def move(self):
        pass

class Dog(Animal):
    def move(self):
        print('멍멍멍')
        
class Bird(Animal):
    def move(self):
        print('짹짹짹')
        
class Tree:
    pass
````

````python
array = [Dog(), Bird(), Tree()]

for x in array:
    if (isinstance(x,Animal)):
        x.move()
````
