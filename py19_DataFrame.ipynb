```python
# ex4. 카테고리 타입 변수(컬럼)의 빈도수를 출력

print(tips.sex.value_counts())
print(tips.smoker.value_counts())
print(tips.day.value_counts())
print(tips.time.value_counts())
```

```python
# ex5. 영수증 금액이 최댓값인 레코드(행)을 출력
tips[tips.total_bill == tips.total_bill.max()]
```

```python
# ex5-2. 영수증 금액이 최솟값인 레코드를 출력
tips[tips.total_bill == tips.total_bill.min()]
```

```python
# ex6. 영수증 금액이 최댓값이거나 최솟값인 레코드들을 출력

tips[(tips.total_bill == tips.total_bill.max()) | 
     (tips.total_bill == tips.total_bill.min())]
```

```python
tips.iloc[[tips.total_bill.argmax(), tips.total_bill.argmin()]]
```

```python
# ex7. 남성 & 여성의 tip 평균

xy = tips[tips.sex == 'Male'].tip.mean()
xx = tips[tips.sex == 'Female'].tip.mean()

print('남자 평균 : ', xy)
print('여자 평균 : ', xx)
```

```python
# 성별 컬럼의 unique(중복되지 않는 값)
tips.sex.unique()
```

```python
for s in tips.sex.unique():
    print(s, ':', tips[tips.sex == s]['tip'].mean())
```

```python
# ex8. 남성 & 여성의 tip 최댓값 / 최솟값 / 중앙값

xy_max = tips[tips.sex == 'Male'].tip.max()
xx_max = tips[tips.sex == 'Female'].tip.max()

print('남자 최댓값 : ', xy_max)
print('여자 최댓값 : ', xx_max)

xy_min = tips[tips.sex == 'Male'].tip.min()
xx_min = tips[tips.sex == 'Female'].tip.min()

print('남자 최솟값 : ', xy_min)
print('여자 최솟값 : ', xx_min)

xy_median = tips[tips.sex == 'Male'].tip.median()
xx_median = tips[tips.sex == 'Female'].tip.median()

print('남자 중앙값 : ', xy_median)
print('여자 중앙값 : ', xx_median)
```

```python
for s in tips.sex.unique():
    print(s, 'max :', tips[tips.sex == s]['tip'].max())
for s in tips.sex.unique():
    print(s, 'min :', tips[tips.sex == s]['tip'].min())
for s in tips.sex.unique():
    print(s, 'median :', tips[tips.sex == s]['tip'].median())
```

```python
# ex+1. 흡연 여부별로 팁의 차이?
    # 흡연 여부별 팁의 평균

for s in tips.smoker.unique():
    print(s, ':', tips[tips.smoker == s].tip.mean())
```

```python
tips.groupby(by='smoker').tip.mean()
```

```python
# ex+2. 요일별 팁의 차이?

for d in tips.day.unique():
    print(d, ':', tips[tips.day == d].tip.mean())
```

```python
tips.groupby(by='day').tip.mean()
```

```python
# ex+3. 시간별 팁의 차이?

for t in tips.time.unique():
    print(t, ':', tips[tips.time == t].tip.mean())
```

```python
tips.groupby(by='time').tip.mean()
```

```python
# 성별, 흡연여부별 팁의 평균

tips.groupby(['sex','smoker']).tip.mean()
```

```python
# 성별, 요일별 팁의 평균

tips.groupby(by= ['sex','day']).tip.mean()
```

```python
tips.groupby(by= ['day','sex']).tip.mean()
```

## 카테고리별로 2개 이상의 컬럼을 선택

```python
# 성별 영수증 금액의 평균, 팁의 평균

tips.groupby('sex')[['total_bill','tip']].mean()
```

```python
# 성별, 흡연여부별 영수증 금액의 평균, 팁의 평균

tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].mean()
```

## 그룹별 2개 이상의 통계 함수 적용

* aggregate 

```python
# 성별 팁의 평균, 최댓값, 최솟값

tips.groupby('sex').tip.agg([np.mean, np.max, np.min]) # aggregate
```

```python
# 성별 흡연여부별 영수증 금액과 팁의 평균, 표준편차
tips.groupby(['sex','smoker'])[['total_bill', 'tip']].agg([np.mean, np.std]) # aggregate
```
