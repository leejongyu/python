#`Print` 함수
백팃[``]으로 감싸면 코드의 형식으로 보이게 됨

##`print()` 함수에 여러개의 argument를 전달하면, argument들을 공백으로 연결해서 한 줄에 출력함.

```python
print('hello python', '안녕 파이톤!')
```

##print() 함수에서 `sep`를 이용해서 공백이 아닌 값으로 연결/ 구분해서 출력함.

```python
print('hello python', '안녕 파이톤!', sep='::')
```

#메서드를 사용한 출력

##`format()` 메서드를 사용한 출력

```python
print('{} + {} + {} = {}')
```

```python
print('{} + {} + {} = {}'.format(1,3,5,9))
```

```python
print('{} + {} + {} = {}'.format('테', '레', '비', '테레비'))
```

```python
print('{x} + {y} + {z} = {xyz}')
```

```python
print('{w} + {x} + {y} = {z}'.format(w=1, x=3, y=5, z=9))
```

```python
print('{w} + {x} + {y} = {z}'.format(w='테', x='레', y='비', z='테레비'))
```

## `formatted string`을 사용한 출력
변수를 이용한 출력

```python
a = 10
b = 20
```

```python
print(f' {a} + {b} = {a+b}')
```

#코드셀 실행 방법

* 코드셀 왼쪽의[실행] 버튼 클릭
* Ctrl + enter : 현재 포커스가 있는 셀을 실행.
* Ctrl + shift + enter : 현재 포커스가 있는 셀을 실행 + 다음 셀로 포커스를 이동.
    * 만약, 현재 셀 아래에 다음 셀이 없는 경우에는 새로운 코드 셀을 생성.
* Ctrl + F9 : 모든 셀 실행.
* Ctrl + F8 : 이전 셀 실행.
* Ctrl + F10 : 다음 셀 실행.
