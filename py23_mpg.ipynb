# mpg (Mile Per Gallon) 데이터 셋

* mpg 데이터 셋을 로드하고 탐색.
* mpg ~ displacement scatter plot
* 2x2 subplot을 생성. scatter plot.
    * mpg ~ cylinders
    * mpg ~ horsepower
    * mpg ~ weight
    * mpg ~ acceleration
* cylinders별 mpg 평균 막대 그래프
* cylinders별 mpg box plot
* origin별, cylinders별 mpg box plot
* origin별, cylinders별 mpg 평균 막대 그래프

```python
import numpy as np # numpy.py 안에 있는 함수를 이용
import pandas as pd 
import matplotlib.pyplot as plt # matplotlib - pyplot.py 안에 있는 함수를 이용
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
mpg.tail()
```

```python
mpg.info()
```

```python
mpg.describe()
```

```python
mpg.origin.value_counts()
```

```python
sns.countplot(data = mpg, x = mpg.origin)
```

```python
sns.boxplot(data = mpg, y = mpg.mpg)
```

```python
# mpg ~ displacement scatter plot
sns.scatterplot(data=mpg, y = mpg.mpg, x= mpg.displacement)
plt.show()
```

```python
sns.scatterplot(data=mpg, y = mpg.mpg, x= mpg.displacement, hue='origin')
plt.show()
```

```python
#2x2 subplot을 생성. scatter plot.

fig, ax = plt.subplots(2,2, figsize = (10,10))

sns.scatterplot(data= mpg, y = mpg.mpg, x = mpg.cylinders, ax = ax[0][0])
sns.scatterplot(data= mpg, y = mpg.mpg, x = mpg.horsepower, ax = ax[0][1])
sns.scatterplot(data= mpg, y = mpg.mpg, x = mpg.weight, ax = ax[1,0])
sns.scatterplot(data= mpg, y = mpg.mpg, x = mpg.acceleration, ax = ax[1,1])

plt.show()
```

```python
# cylinders별 mpg 평균 막대 그래프

sns.barplot(data = mpg, x = 'cylinders', y = 'mpg')
plt.show()
```

```python
# cylinders별 mpg box plot


sns.catplot(data = mpg, x = 'cylinders', y = 'mpg')
plt.show()
```

```python
# origin별, cylinders별 mpg box plot
sns.catplot(data = mpg, x = 'origin', y = 'mpg', col='cylinders', kind='box')
```

```python
# origin별, cylinders별 mpg 평균 막대 그래프

sns.catplot(data = mpg, x = 'cylinders', y = 'mpg', col='origin', kind='bar', estimator=np.mean)
```
