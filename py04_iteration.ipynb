# for 반복문
```
for 변수 in iteravle_type:
    반복해서 실행할 코드 블록
```


##iterable_type : 반복 가능한 타입의 객체 
    : range, list, tuple, dict, set, srt, ....
* `range` 함수
    * `range(stop)` :  0 <= x < stop 범위의 1씩 증가하는 정수들을 생성.
    * `range(start, stop)` : start <= x < stop 범위의 1씩 증가하는 정수들을 생성.
    * `range(start, stop, step)` : 
        * step > 0 인 경우 : start <= x < stop 범위의 step만큼 씩 증가하는 정수들을 생성
        * step < 0 인 경우 : start >= x > stop 범위의 step만큼 씩 감사하는 정수들을 생성.

````python
for x in range(5): # 0 <= x < 5
    print(x, end='')
````

````python
for x in range(1,6): # 1 <= x < 6
    print(x, end='')
````

````python
for x in range(0,100,5): # 0 <= x < 100 (0부터 5씩 증가)
    print(x, end=', ')
````

````python
for x in range(100,0,-5) : # 100 >= x > 0 (100부터 5씩 감소)
    print(x, end=',')
````

# 중첩 반복문


````python
#구구단 2단부터 9단까지 출력
for x in range(2,10): # 2 <= x < 10
    for y in range(1, 10): # 1 <= x < 10
        print(f'{x} x {y} = {x*y}')
    print('-'*20)
````

# Break와 continue

*   `break`: iteration을 종료하고, break가 포함된 가장 가까운 반복문을 빠져 나옴.
*   `continue`: 반복문 내부의 실행을 멈추고, 그 다음 iteration을 계속함.


````python
for x in range(5):
    if x == 2:
        break
    print(x)
````

````python
for x in range(5):
    if x == 2:
        continue
    print(x)
````

## 구구단 해당 단수까지만 출력

````python
for x in range(2,10): # 2 <= x < 10
    for y in range(1, 10): # 1 <= x < 10
        print(f'{x} x {y} = {x*y}')
        if x == y:
            break
    print('-'*20)
````

````python
for x in range(2,10): # 2 <= x < 10
    for y in range(1, x+1): # 1 <= x < 10
        print(f'{x} x {y} = {x*y}')
    print('-'*20)
````

# while 반복문
```
[초기화 문장]
while 조건식:
    조건식이 참일 때 실행할 코드 블록.
    [조건을 변경할 수 있는 문장] => 생략 가능한 문장
```


````python
x = 1
while x <= 5:
    print('*'* x)
    x += 1
    if (x==5):
        break
````

````python
x = 1 
while x <= 10:
    print(x, end=',')
    x+=2
````
