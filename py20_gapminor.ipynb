```python
# 1. GitHub에 저장된 gapminder.tsv 파일을 읽고 DataFrame을 생성
# csv(comma-Separated Values)
# tsv(tap-Separated Values)
file_path = 'https://github.com/JakeOh/202208_itw_java134_lab_python/raw/main/gapminder.tsv'
exer = pd.read_csv(file_path, sep ='\t')
```

```python
# 2. 데이터 프레임 탐색(확인)
exer.head()
```

```python
exer.tail()
```

```python
exer.info()
```

```python
# 3. (중복되지 않는) 국가 이름의 개수?

len(exer.country.unique())
```

```python
exer['country'].value_counts()
```

```python
# 4. (중복되지 않는) 대륙 이름?

exer.continent.unique()
```

```python
exer['continent'].value_counts()
```

```python
# 5. (중복되지 않는) 연도의 개수?

len(exer.year.unique())
```

```python
# 6. 국가이름(country), 연도(year), 1인당 GDP 컬럼(gdpPercap)을 선택

exer[['country', 'year', 'gdpPercap']]
```

```python
# 7-1 . 1인당 GDP의 최댓값을 찾고, 그 row를 출력
exer[exer.gdpPercap == exer.gdpPercap.max()]
```

```python
exer.gdpPercap.max()
```

```python
# 7-2 . 1인당 GDP의 최솟값을 찾고, 그 row를 출력
exer[exer.gdpPercap == exer.gdpPercap.min()]
```

```python
exer['gdpPercap'].min()
```

```python
exer[(exer.gdpPercap == exer.gdpPercap.max()) |
     (exer.gdpPercap == exer.gdpPercap.min())]
```

```python
# 8-1. 기대수명(lifeExp)의 최댓값을 찾고, 그 행들을 출력
exer[exer.lifeExp == exer.lifeExp.max()]
```

```python
# 8-2. 기대수명(lifeExp)의 최솟값을 찾고, 그 행들을 출력
exer[exer.lifeExp == exer.lifeExp.min()]
```

```python
exer[(exer.lifeExp == exer.lifeExp.max()) |
     (exer.lifeExp == exer.lifeExp.min())]
```

```python
# 일본의 연도별 기대 수명
exer[exer.country == 'Japan'][['country', 'year', 'lifeExp']]
```

```python
# 르완다의 연도별 기대 수명
exer[exer.country == 'Rwanda'][['country', 'year', 'lifeExp']]
```

```python
# 9. 연도별 기대수명의 평균을 출력
for y in exer.year.unique():
    print(y, ':' , exer[exer.year == y].lifeExp.mean())
```

```python
exer.groupby('year').lifeExp.mean()
```

```python
# 10. 연도별 1인당 GDP의 평균을 출력
for y in exer.year.unique():
    print(y, ':' , exer[exer.year == y].gdpPercap.mean())
```

```python
exer.groupby('year').gdpPercap.mean()
```

```python
# 11. 연도별 대륙별 전세계 인구를 출력

for y in exer.year.unique(): 
    for x in exer.continent.unique():
        print(y, 'years,', x, ':' , exer[exer.year ==y][exer.continent == x]['pop'].sum())
```

```python
exer.groupby(['year','continent'])[['pop']].sum()
```
