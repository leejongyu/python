#set(집합)

* 중복된 데이터를 허용하지 않음
* 데이터 저장 순서가 중요하지 않음
    * 인덱스(index)가 없음 -> indexing, sliding 기능을 제공하지 않음.
* [] = list, () = tuple, {} = set

```python
s = {1,2,3,1,2,3,3,2,1}

print(s)
```

```python
for x in s:
    print(x, end=' ')
```

#set의 기능(메서드)

* `set.add(value)` : value값 추가
* `set.remove(value)` : value값 삭제
* `set1.union(set2)` : set1과 set2의 합집합
* `set1.intersection(set2)` : set1과 set2의 교집합
* `set1.difference(set2)` : set1과 set2의 차집합

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}

print(s1.union(s2))
print(s1 | s2) # 합집합 연산자
print(s1.intersection(s2))
print(s1 & s2) # 교집합 연산자
print(s1.difference(s2))
print(s1 - s2) # 차집합 연산자
```
