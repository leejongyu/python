# List

list = 여러개의 값들을 하나의 변수에 저장하기 위한 데이터 타입.

* element(원소, 요소): 리스트에 저장되는 각각의 값.
* index(인덱스): 리스트에 원소들이 저장된 위치.
    * 리스트의 인덱스는 **0**부터 시작.
    * 리스트의 마지막 인덱스는 **(전체 원소의 개수 -1)**
* 리스트는 원소들을 추가/ 삭제/ 변경 가능.
    * **`tuple : 값들을 변경할 수 없는 리스트`**과 다른점!


````python
# 숫자들을 저장하는 리스트
numbers = [10, 20 ,30, 8,456,7,9,50,9,3,1,1]

print(numbers)
````

## indexing

인덱스를 사용해서 리스트에서 원소를 참조하는 방법.

* indexing의 결과는 원소 1개의 값

````python
print(numbers[6])
print(numbers[-1])

print(numbers[5])
print(numbers[-2])
````

## slicing

인덱스를 사용해서 리스트의 부분집합(리스트)를 잘라내는 방법.

```
list[start:end]
```
* start <= index < end 범위의 인덱스 위치의 원소들을 잘라냄.
* start를 생략하는 경우에는 첫번째 원소부터 잘라냄.
* end를 생략한 경우에는 마지막 원소까지 잘라냄.
* slicing의 결과는 새로운 리스트를 생성

````python
# 리스트 numbers에서 첫번째 원소부터 순서대로 3개 잘라냄

slicing1 = numbers[:3]

print(slicing1)

# 리스트 numbers에서 끝에서 3새의 원소를 잘라냄

slicing2 = numbers[-3:]
print(slicing2)
````

#list의 기능

##list의 산술 연산

* `list + list`
* `list * 정수` or `정수 * list`

##list 객체의 기능(메서드)

* 추가 기능
    * `list.append(value)` : 리스트의 가장 마지막에 새로운 원소(value)를 추가.
    * `list.insert(index,value)` : 리스트의 index 위치에 value를 추가.
    * `list.extend(iterable)` : list + list2(iterable)

* 삭제 기능
    * `list.remove(value)` : 리스트에서 처음 등장하는 value를 삭제.
    * `list.pop(index)` : 리스트레서 index 위치에 있는 원소를 삭제.
    * `list.clear()` : 리스트의 모든 원소를 삭제.

## Ex 1.
1. 비어있는 list를 생성
2. [1, 10)의 정수 난수 5개를 저장
3. 리스트 원소들의 합을 출력
4. 리스트 원소들의 평균을 출력

````python
list1 = []

for x in range(5):
    list1.append(random.randrange(1,10))

print(list1)

sum = 0

for x in list1:
    sum += x

print('sum', ':', sum)

avg = sum/len(list1)

print(avg)
````

## Ex 2.
1. 비어있는 list를 생성
2. [10, 100]의 정수 난수 10개를 저장
3. 짝수들만 저장된 evens 리스트를 생성
4. 홀수들만 저장된 odds 리스트를 생성

````python
list2 = []

for _ in range(10):
    list2.append(random.randint(10,100))

print(list2)


odds1 = []
evens1 = []

for x in list2:
    if (x%2):
        odds1.append(x)
    else:
        evens1.append(x)

print(odds1)
print(evens1)
````

## Ex 3.
1. 비어있는 list를 생성
2. [-10, 10]의 정수 난수 10개를 저장
3. 원소가 양수이면 True, 그렇지 않으면 False를 저장하는 리스트 생성

````python
list3 = []

for x in range(10):
    list3.append(random.randint(-10,10))

print(list3)

TorF=[]
for x in list3:
    if x >0:
        TorF.append([x, 'True'])
    elif x == 0:
        TorF.append([x, 'Lucky'])
    else:
        TorF.append([x, 'False'])

print(TorF)
````

##Ex 4.

2. gender_codes의 값이 0이면 'M' else 'F'를 저장하는 리스트 생성

#List comprehension(리스트 내포)

````python
numbers = [x for x in range(1, 11)]

print(numbers)
````

````python
# ex1

list1 = [random.randrange(1,11) for _ in range(5)]

print(list1)

sum = 0

for x in list1:
    sum+=x
    
print(sum)

sum/len(list1)
````

````python
# ex2
list2 = [random.randint(10,100) for _ in range(1,11)]

print(list2)

evens = [x for x in list2 if x % 2 == 0]

print(evens)

odds = [x for x in list2 if x % 2]

print(odds)

if (len(odds) == len(evens)):
    print('lucky')
````

````python
# ex3
list3 = [random.randint(-10,10) for x in range(1,11)]

print(list3)

is_true = [x > 0 for x in list3]
is_positive = [True if x > 0 else False for x in list3]
print(is_positive)
print(is_true)
````

# 과제 
* 1. 0 이상 10 미만의 정수 난수 10개를 저장하는 리스트 numbers1을 만들고 출력
* 2. 0 이상 10 미만의 정수 난수 10개를 저장하는 리스트 numbers1을 만들고 출력
* 3. numbers1과 numbers2에서 같은 인덱스의 원소들끼ㅣㄹ 합을 저장하는 리스트 adds를 만들고 출력.
* 4. numbers1의 원소에서 numbers2의 원소를 뺀 값들을 저장하는 리스트 subtract를 만들고 출력.
* 5. numbers1과 numbers2에서 같은 원소들끼리의 곱을 저장하는 리스트 multiplications를 만들고 출력

````python
numbers1 = [random.randint(0,10) for _ in range(1,11)]
numbers2 = [random.randint(0,10) for _ in range(1,11)]

print(numbers1)
print(numbers2)

adds = []
subtracts = []
multiplications = []

for xi, xv in enumerate(numbers1):
    for yi,yv in enumerate(numbers2):
        if (xi == yi):
            adds.append(xv+yv)
            subtracts.append(xv-yv)
            multiplications.append(xv*yv)
            
print(adds)
print(subtracts)
print(multiplications)
````

````python
numbers3 = [random.randint(0,10) for _ in range(1,11)]
numbers4 = [random.randint(0,10) for _ in range(1,11)]

print(numbers3)
print(numbers4)

adds2 = []


adds2 = [numbers3[i] + numbers4[i] for i in range(len(numbers3))]

print(adds2)
````

````python
print(numbers3)
print(numbers4)

subtracts2 = [x - y for x,y in zip (numbers3,numbers4)]
multiplies2 = [x * y for x,y in zip (numbers3,numbers4)]

print(subtracts2)
print(multiplies2)
````
