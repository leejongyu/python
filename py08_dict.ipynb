#dict

* dictionary(사전) 형식의 데이터 타입.
* 키(key)를 기반으로 값(value)를 저장하는 데이터 타입.
    * list, tuple : 인덱스(index)를 기반으로 값을 저장하는 데이터 타입.
    * dict에서 키(key)의 역할은 리스트나 튜플에서 인덱스와 비슷한 역할 : dict의 원소 1개를 참조하기 위한 용도.
    * dict에서 key를 사용하는 방법은 list나 tuple과 마찬가지로 index 연산자([])를 사용

```python
students={1:'홍길동', 2:'김길동', 3:'이길동'}

print(students)

print(students[1])

students[4] = '박길동' #원소 추가 : dict[new_key] = new_value

print(students)

students[2] = '최길동' #원소 변경 : dict[old_key] = new_value

print(students)

students.pop(3) #원소 삭제 : dict.pop(old_key)

print(students)
```

## dict에서 key와 value는 다양한 타입의 자료들을 사용할 수 있음.

```python
contact= {
    'no' : 1, 
    'name' : '오쌤', 
    'phone':['010-0000-0000', '02-000-0000'],
    'email':{'com': 'lee@naver.com', 'personal':'hello@naver.com'}
}

print(contact)
print(contact['no'])
print(contact['phone'])
print(contact['phone'][0])
print(contact['email'])
print(contact['email']['com'])
```

#dict와 for반복문

for-in 구문에서 dict 타입 객체를 사용하면 key에 대해서 반복(iteration)함.

```python
for k in contact:
    print(k)
```

```python
for k in contact:
    print(k, ':' , contact[k]) 
```

```python
contact.items() # (key,value)

for k,v in contact.items():
    print(k,':',v)
```

#Dictionary Comprehension

```python
emp_no = [1000,2000,1001,2001]
emp_name = ['king', 'scott', 'allen', 'lee']

emp = {}

for x,y in zip(emp_no, emp_name):
    emp[x] = y

print(emp)
```

```python
emp2 = {k:v for k,v in zip(emp_no, emp_name)}

print(emp2)
```

```python
messages = ['Happy New Year!', 'Merry ChristMas!', '새해 복 많이 받으세요']

message_lengths = {}

for k in messages:
    message_lengths[k] = len(k)

print(message_lengths)

mes = {v:len(v) for v in messages}

print(mes)
```
