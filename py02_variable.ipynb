# 파이썬 자료 타입

파이썬은 변수를 선언할 때 자료 타입을 명시하지 않음! 

변수의 자료 타입은 코드가 실행 될 때 결정됨.

* 숫자 타입
    * `int` : 정수
    * `float` : 실수

* 문자열 타입 
    * `str` : 문자열

* 논리 타입
    * `bool` : True / False (대문자로 시작함)

* 객체 타입


```python
# 주석(comment)
# 변수_이름 = 값
n = 13

print(n)
print(type(n), n) # type(arg) : arg의 자료 타입을 리턴.

n = 3.14

print(n)
print(type(n), n) # type(arg) : arg의 자료 타입을 리턴.

n = '글자'

print(n)
print(type(n), n) # type(arg) : arg의 자료 타입을 리턴.
```

식(expression) : 값, 변수, 계산식(연산), 함수 호출.

문장(statement) : cord가 실행되기 위한 최소 실행 단위. 

ipynb(python 노트북)의 코드셀에서 가장 마지막 줄에 있는 문장이 식(expression)이 있으면 자동으로 출력

```python
 x = 1 # 문장
 y = 2 # 문장
 x + y # 식 
```

## 명시적 타입 변환

* `int(값)` : 값을 정수 타입으로 변환해서 리턴
* `float(값)` : 값을 실수 타입으로 변환해서 리턴
* `str(값)` : 값을 문자열로 변환해서 리턴.

```python
# colab 노트북의 콘솔에서 입력 받는 방법:

n = input('정수 입력 : ')

print(type(n)) # n = '100') 이라는 문자열.

print(n) 

n = int(n) # string '100' 을 int 100으로 명시적 타입 변환

print(type(n))

print(n)
```

# 산술 연산자

종류 : `+` , `-`, `*`, `/`, `%`,`//`, `**`

* `/`  : 소수점까지 계산하는 나눗셈
* `//` : 나눈 몫만 계산하는 나눗셈
* `%`  : 나눈 나머지만 계산하는 나눗셈
* `**` : 거듭 제곱

```python
print(10/3)

print(10//3)

print(10%3) 
```

```python
10 ** 1
```

# 20221223 과제1. 조건문

* 3개의 정수를 입력 받고 각각 변수에 저장
* 3개 정수의 평균을 소수점까지 계산하고 출력.
* 평균이 90점 이상이면 A, 80 이상이면 B, 70점 이상이면 C , 70점 미만이면 F 출력


```python
a = input('점수 입력 : ')
b = input('점수 입력 : ')
c = input('점수 입력 : ')

a = int(a)
b = int(b)
c = int(c)

x = (a+b+c)/3

print(x)

if x >= 90 :
    print('A')
elif 80 <= x < 90 :
    print('B')
elif 70 <= x < 80 :
    print('C')
else:
    print('F')
```
