#`str` 클래스의 메서드

* `str` 클래스의 메서드들은 원본 문자열을 변경하지 않음!
* 원본 문자열을 변경한 새로운 문자열을 리턴


```python
s = 'pyThOn 안녕 PyThOn'

s
```

```python
s.lower()
```

```python
s.upper()
```

```python
s.capitalize() # 첫글자만 대문자로, 나머지는 소문자로 변경
```

```python
s.count('p') # 문자열에서 substring이 등장하는 횟수를 리턴
```

```python
url.startswith('ht')
```

```python
phone = '010-0000-0000'
```

```python
phone.split(sep='-')
```

```python
sentence = 'hellp python \t 안녕하세요 \n Java'

print(sentence)

sentence.split() # 파라미터 sep=None 인 경우, 공백(white space: 스페이스, 탭, 줄바꿈)을 기준으로 자름.
```

```python
#CSV(comma-seperated calues) 파일 : 데이터들이 comma(,)로 구분된 형식의 파일 

csv = '100, scott, 200, scott@oracle.com, 1234.56, IT, 2022-12-29'

csv2 = csv.split(',')
```

```python
print(csv2)

','.join(csv2)
```

```python
s2 = '                           HelLo wOrrldO.... 하. ㅎ핳ㅎ.하..                      '

print(s2)
print(s2.lstrip()) #문자열의 왼쪽에 있는 공백 제거
print(s2.rstrip()) #문자열의 오른쪽에 있는 공백 제거
print(s2.strip()) #문자열의 양 끝(rt, lt)에 있는 공백 제거
```
