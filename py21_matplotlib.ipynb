#`matplotlib.pyploy` 모듈을 사용한 시각화

1. `figure` 객체 생성
2. `figure` 객체에 subplpt(axis)을 생성해서 추가 
3. 2번에서 생성된 subplot에 그래프 그림.
4. 그래프의 여러가지 옵션들을 설정.
5. 그래프 보여주기.

```python
# 1. figure 객체 생성
fig = plt.figure() # 빈 스케치북 만들기

# 2. fig 객체에 subplot을 생성해서 추가
axis = fig.add_subplot(1,1,1) # 스케치북을 나누기(row, column, index). add_subplot의 index는 1부터 시작

# 3. subplot에 그래프종류를 선택해서 그림.
axis.hist(x=tips['total_bill'])

# 4. 그래프의 옵션 설정
axis.grid()
axis.set_xlabel('Total_bill')
axis.set_ylabel('Count')
axis.set_title('Histogram')

# 5. 그래프 보여주기
plt.show()
```

## 2개의 subplot을 상하로 배치하고 그래프를 그리기.

```python
#1. figure 생성
fig = plt.figure(figsize=(8,8)) # figuresize=(width, height)

#2. subplot 추가
ax1 = fig.add_subplot(2,1,1) #add_subplt(211)
ax2 = fig.add_subplot(212) #add_subplt(2,1,2)와 동일한 코드

#3. 각 subplot에 그래프 그림.
ax1.hist(x=tips.total_bill) # 히스토그램
ax2.scatter(x=tips.total_bill, y= tips.tip) # 산점도

#4. 옵션 설정.
ax1.set_title('Histogram')
ax2.set_title('Scatter Plot')
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')

#5. 보여주기
plt.show()
```

# Box plot(상자 그림)

숫자타입 변수(컬럼)에 기술 통계량(descriptive statistics)을 상자 모양으로 시각화.

* 최솟값, 최댓값, 4분위값(25%, 50%, 75%)

* 박스 = 전체 데이터의 50%가 들어있는 구간. 데이터들의 분산을 보기 위한 그래프
    * 박스 크기가 크다? 전체 데이터의 50%가 넓은 구간에 분포
    * 박스 크기가 좁다? 전체 데이터의 50%가 좁은 구간에 분포

* 수염/ 펜스 : (박스크기 = 75% - 25%) * 1.5
    * 수염 바깥으로 데이터가 더 있으면, 끝에 펜스를 치고 데이터를 표시 (outlayer)
    * 수염 안쪽에만 데이터가 있으면, 그 부분까지만 펜스를 침

```python
# tips 데이터 프레임에서 total_bill 변수(컬럼)의 기술 통계량

print(tips.total_bill.describe())

plt.boxplot(x=tips.total_bill) 
plt.grid()

plt.show()
```

```python
# tips 데이터 프레임에서 tip 변수(컬럼)의 기술 통계량

print(tips.tip.describe())

plt.boxplot(x=tips.tip) 
plt.grid()

plt.show()
```

# Histogram (히스토그램)

연속형 자료를 구간으로 나눠서 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.

```python
plt.hist(x=tips.total_bill, bins=20, edgecolor='black', color='green')
# x : 히스토그램을 그리기 위한 자료(변수)
# bins : 막대 개수 또는 구간
# color : 막대 안쪽을 채우는 색깔
# edgecolor : 막대 테두리 선 색깔

plt.ylabel('Count')
plt.xlabel('Total bill')
plt.grid()
plt.show()
```

```python
plt.hist(x=tips.tip, bins=20, edgecolor='black', color='gray')
# x : 히스토그램을 그리기 위한 자료(변수)
# bins : 막대 개수 또는 구간
# color : 막대 안쪽을 채우는 색깔
# edgecolor : 막대 테두리 선 색깔

plt.ylabel('Count')
plt.xlabel('Tip')
plt.grid()
plt.show()
```

# Scatter plot(산정도)

두 변수 사이의 상관관계를 파악하기 위하여 사용

`y ~ x`. `y = ax +b` ...

```python
# tips ~ total_bill

plt.scatter(x=tips.tip, y=tips.total_bill, alpha=0.3, c='blue', s=100, marker ='*')

# alpha(0 ~ 10: 0 = 완전 투명한 색) = 불투명도
# c = 마커의 색 (color)
# s = 마커의 크기 (size)
# marker = 마커의 모양 ('s = square', '^ = 삼각형' , 'v = 역삼각형', '* = 별모양' ....)

plt.xlabel('Tip')
plt.ylabel('Total bill')

plt.show()
```

```python
# tip ~ size : boxplot과 같이 데이터의 분산을 보여줌

plt.scatter(x=tips['size'], y=tips['tip'], alpha=0.4)
```

# Line plot(선 그래프)

주로 시계열 데이터 (시간에 따라서 변하는 데이터: 주가, 환율, 날씨 ...)를 시각화

```python
np.random.seed(1)

df = pd.DataFrame(data = {'x1' : np.arange(11,21), 'x2' : np.random.rand(10)})

df
```

```python
# x2 ~ x1 (y ~ x)
plt.plot(df.x1, df.x2, 'g+--')
# plt.plt([x], y, [fmt])
# [fmt = format] : 문자열 , '[선, 마커의 색], [마커의 모양], [선의 스타일]'
# 선, 마커의 색 = b(lue), r(ed), (blac)k, g(reen), ...
# 마커 모양 : o, s ,^ ,v, *, +, ...
# 선 스타일 : -(직선, solid), :(점선, dotted), --(긴점선, dashed)

plt.grid()

plt.show()
```

```python
# x2 ~ index

plt.plot(df.x2, c= 'red', marker = 'v', linestyle=":")

plt.grid()
plt.show()
```

## ex3

    * scatter plot (y = tip, x = total_bill)
    * 요일별로 점의 색깔, 모양을 다르게 표현
    * 범례를 표시

```python
days = tips.day.unique()

for d in days:
    data = tips[tips.day == d]
    if d == 'Sun':
        c = 'r'
        m = 'o'
    if d == 'Sat':        
        c = 'b'
        m = '*'
    if d == 'Thur':
        c = 'g'
        m = '+'
    if d == 'Fri':
        c = 'orange'
        m = '^'

    plt.scatter(x = data.total_bill, y = data.tip, alpha = 0.5 ,label = d, color = c, marker = m)


plt.ylabel('Tip')
plt.xlabel('Total bill')
plt.legend()

plt.show()
```

```python
days = tips.day.unique()

markers = ['o', '+', 's', 'v']
colors = ['r', 'g', 'b', 'orange']
count = 0
for d in days:
    data = tips[tips.day == d]
    plt.scatter(x = data.total_bill, y = data.tip, alpha = 0.5 ,label = d, color = colors[count], marker = markers[count])
    count += 1

plt.ylabel('Tip')
plt.xlabel('Total bill')
plt.legend()

plt.show()
```
